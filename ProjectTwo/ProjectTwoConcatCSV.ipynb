```python
import pandas as pd
from glob import glob
```

```python
HVI_files = sorted(glob('datasets/LBhvi_*.csv'))
HVI_files
```

```python
Sqft_files = sorted(glob('datasets/LB_MedianRentalPricePerSqft_*.csv'))
Sqft_files
```

```python
Rent_Unit_files = sorted(glob('datasets/LB_MedianRentalPrice_*.csv'))
Rent_Unit_files
```

```python
HVI_data = pd.concat((pd.read_csv(file, encoding = "ISO-8859-1").assign(filename = file)
          for file in HVI_files), ignore_index = True, sort = False)
```

```python
Sqft_data = pd.concat((pd.read_csv(file, encoding = "ISO-8859-1").assign(filename = file)
          for file in Sqft_files), ignore_index = True, sort = False)
```

```python
Rent_Unit_data = pd.concat((pd.read_csv(file, encoding = "ISO-8859-1").assign(filename = file)
          for file in Rent_Unit_files), ignore_index = True, sort = False)
```

```python
HVI_data
```

```python
long_beach = ['Long Beach']
LB_hvi = HVI_data.loc[HVI_data['RegionName'].isin(long_beach)]
LB_hvi.head()
```

```python
CA_longbeach = ['CA']
LB_hvi = LB_hvi.loc[LB_hvi['State'].isin(CA_longbeach)]
LB_hvi
```

```python
for col in LB_hvi.columns:
    if col <= '2018-12':
        del LB_hvi[col]
```

```python
LB_hvi = LB_hvi.drop(columns=['Metro', 'CountyName'])

LB_hvi
```

```python
LB_hvi = LB_hvi.rename(columns={"2019-01":"January", "2019-02":"February",
                               "2019-03":"March", "2019-04":"April",
                               "2019-05":"May", "2019-06":"June",
                               "2019-07":"July", "2019-08":"August",
                               "2019-09":"September", "2019-10":"October",
                               "2019-11":"November"})
LB_hvi
```
