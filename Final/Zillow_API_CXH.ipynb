# Zillow API Calls for Lat, Lng, Zestimate Value, and link to details.

## Imports

```python
import pandas as pd
import numpy as np
from config import API_KEY
import zillow
import pprint
import time
```

```python
buy_list = pd.read_csv('Data/buy_property_list.csv')
rent_list = pd.read_csv('Data/rent_property_list.csv')
```

### Buy List

```python
add_list=buy_list['address']
updated_add = []
for add in add_list:
    new_add = add.replace("+", " ")
    updated_add.append(new_add)
len(updated_add), updated_add
```

```python
buy_list['address']=updated_add
buy_list
```

```python
api = zillow.ValuationApi()
citystatezip = "Long Beach, CA"
data = api.GetSearchResults(API_KEY, buy_list.address[0], citystatezip)
result = data.get_dict()
pp = pprint.PrettyPrinter(indent=2)
pp.pprint(result)
```

```python
api = zillow.ValuationApi()
citystatezip = "Long Beach, CA"
buy_zipcode = []
buy_lat = []
buy_lng = []
buy_zpid = []
buy_links = []
counter = -1

for address in buy_list['address']:
    counter += 1
    try:
        data = api.GetSearchResults(API_KEY, address, citystatezip)
        result = data.get_dict()
        buy_zipcode.append(result['full_address']['zipcode'])
        buy_lat.append(result['full_address']['latitude'])
        buy_lng.append(result['full_address']['longitude'])
        buy_zpid.append(result['zpid'])
        buy_links.append(result['links']['home_details'])
    except:
        print(f"failed on {address}, number {counter}")
        buy_zipcode.append(np.nan)
        buy_lat.append(np.nan)
        buy_lng.append(np.nan)
        buy_zpid.append(np.nan)
        buy_links.append(np.nan)
    
```

```python
len(buy_zipcode), len(buy_lat), len(buy_lng), len(buy_zpid), len(buy_links)
```

```python
buy_list['zipcode'] = buy_zipcode
buy_list['latitude'] = buy_lat
buy_list['longitude'] = buy_lng
buy_list['zpid'] = buy_zpid
buy_list['link'] = buy_links
buy_list
```

```python
buy_list_new = buy_list.dropna(axis=0, how="any")
buy_list_new.shape
```

```python
buy_list_new.to_csv("Data/buy_full_list.csv")
```
